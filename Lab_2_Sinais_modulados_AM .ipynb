# Bibliotecas

```python
import os
from numpy import *
from matplotlib.pyplot import *
from scipy import signal
import scipy
from scipy.signal import lfilter, windows
```

# Geração do Sinal

```python
# Parâmetros do sinal modulado
Ac = 1  # Amplitude da portadora
fc = 100  # Frequência da portadora em Hz
fs = fc * 10  # Frequência de amostragem.
# Para um sinal analógico, fs > 2 x máxima Frequência do sinal (Nyquist). (Consideramos 10 aqui para que o sinal seja reconstruído em alta resolução)

Ts = 1 / fs  # Calcula o intervalo de tempo.

Am = 1 # pico do sinal modulante em Hz
fm = 10  # Frequência do Sinal Modulante em Hz
T = 4 / fm  # Tempo do sinal gerado (4 períodos).
k = 1  #sensibilidade à amplitude 


# Geração dos vetores do sinal
t = arange(0, T - Ts, Ts)  # Cria vetor tempo.
m = Am * cos(2 * pi * fm * t)  # Criação do sinal modulante
c = Ac * cos(2 * pi * fc * t)  # Criação do sinal da portadora
```

# Traçando no tempo e na frequência

```python
# OBS.: A sua análise será realiza em cima desses sinais destacados. 

# --------------------------------------
# Exemplo de plot do sinal modulado no tempo e na frequência.
# --------------------------------------

# Pode-se utilizar o comando "n.welch".
# Declarando alguns parâmetros:

NFFT = len(t)  # Resolução do espectro computado.
#WINDOW = ones(NFFT)  # Janela de Hamming.
WINDOW = windows.hann(NFFT, False)  # Janela de Hamming.

# OBS.: O entendimento profundo do uso e funcionamento das funções acima não é fundamental para o entendimento do exercício. 

# --- Sinal Modulante----
rcParams['figure.figsize'] = [10, 5]
figure(1)
plot(t, m,'r')  # Onda no domínio do tempo, com cor 'r' red.
# set(gca, 'Xlim', [t[1], t[-1]])
title('Sinal modulante no tempo')
xlabel('Tempo (s)')
ylabel('Amplitude (V)')
grid()

figure(2)
f, p = signal.welch(m, fs, WINDOW,nfft=NFFT , scaling='spectrum')  # Espectro de Frequência (estima a densidade de potência espectral.)
plot(f, p)
title('Espectro de frequência do sinal modulante')
xlabel('Frequência (Hz)')
ylabel('E/Freq (W/Hz)')
grid()
show()



# --- Portadora----
rcParams['figure.figsize'] = [10, 5]
figure(1)
plot(t, c,'r')  # Onda no domínio do tempo, com cor 'r' red.
# set(gca, 'Xlim', [t[1], t[-1]])
title('Sinal modulante no tempo')
xlabel('Tempo (s)')
ylabel('Amplitude (V)')
grid()

figure(2)
f, p_c = signal.welch(c, fs, WINDOW,nfft=NFFT , scaling='spectrum')  # Espectro de Frequência (estima a densidade de potência espectral.)
plot(f, p_c)
title('Espectro de frequência do sinal modulante')
xlabel('Frequência (Hz)')
ylabel('E/Freq (W/Hz)')
grid()
show()
```

# Ex: modulação AM-DSB-SC

```python
# Geração dos vetores do sinal AM-DSB-SC
s_am_dsb_sc = m * c





# --- sinal AM-DSB-SC----
rcParams['figure.figsize'] = [10, 5]
figure(1)
plot(t, s_am_dsb_sc,'r')  # Onda no domínio do tempo, com cor 'r' red.
# set(gca, 'Xlim', [t[1], t[-1]])
title('Sinal modulante no tempo')
xlabel('Tempo (s)')
ylabel('Amplitude (V)')
grid()

figure(2)
f, p_s_am_dsb_sc = signal.welch(s_am_dsb_sc, fs, WINDOW,nfft=NFFT , scaling='spectrum')  # Espectro de Frequência (estima a densidade de potência espectral.)
plot(f, p_s_am_dsb_sc)
title('Espectro de frequência do sinal modulante')
xlabel('Frequência (Hz)')
ylabel('E/Freq (W/Hz)')
grid()
show()
```
